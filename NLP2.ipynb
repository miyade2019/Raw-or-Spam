```python
df['Text']=df['Text'].str.lower()
df['Text'] = df['Text'].str.replace(r'[^\w\s]', '', regex=True)
df['Text']=df['Text'].str.replace('\n','')
df['Text']=df['Text'].str.replace('\d+','')# d = dijital (rakamlar)
df['Text'] = df['Text'].str.replace('\r', '')
```

```python

x=df["Text"]
y=df['Class']
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from textblob import TextBlob
from nltk.stem import PorterStemmer
pr=PorterStemmer()
from sklearn.feature_extraction.text import CountVectorizer
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#ekler kaldırıp kökler bulmma fonksiyoun
def lemmafn(text):
    words=TextBlob(text).words
    return[pr.stem(word) for word in words]
```

```python
vect=CountVectorizer(stop_words="english",ngram_range=(1,2),max_features=10000,analyzer=lemmafn)
```

```python
x=vect.fit_transform(x)
```

```python
#x_train=vect.fit_transform(x_train)
#x_test=vect.fit_transform(x_test)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.20)
```

```python
nb=MultinomialNB()
```

```python
model=nb.fit(x_train,y_train) # modeli eğitimi
```

```python
tahmin=model.predict(x_test)
```

```python
accuracy_score(tahmin,y_test)
```
